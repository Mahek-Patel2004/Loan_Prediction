K-nearest neighbor model implementation

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns="loan_status" , axis=1)
y = df["loan_status"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=21)
```

```python
print('Training dataset shape: ',X_train.shape)
print('Testing dataset shape: ',X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=9  #no of neighbors
knn_classifier = KNeighborsClassifier(n_neighbors=k)
knn_classifier.fit(X_train, y_train)
```

```python
y_pred = knn_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
print('\n\tTrue Positives,TP =  q = ',conf_matrix[0][0])
print('\n\tTrue Negatives,TN =  r = ',conf_matrix[0][1])
print('\n\tFalse Positives,FP =  s = ',conf_matrix[1][0])
print('\n\tFalse Negatives,FN =  t = ',conf_matrix[1][1])
```

Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy*100)
```

```python
loan_status_pred = ['Approved' if pred == 1 else 'Rejected' for pred in y_pred]

# Display the predicted loan statuses for the test set
for i, status in enumerate(loan_status_pred):
    print(f"Loan {i+1}: {status}")
```

```python
import pandas as pd

def predict_loan_approval(model, input_data):
    """
    Function to predict loan approval based on input parameters.
    
    Parameters:
    - model: Trained model (e.g., KNeighborsClassifier).
    - input_data: A dictionary of input features with feature names as keys.
    
    Returns:
    - "Approved" or "Rejected" based on the prediction.
    """
    # Create a DataFrame with the same feature names as in the training data
    input_df = pd.DataFrame([input_data])
    
    # Make the prediction
    prediction = model.predict(input_df)[0]
    
    # Interpret the prediction
    return "Approved" if prediction == 1 else "Rejected"

# Example usage:
# The input data should include the exact feature names that were used to train the model.
input_data = {
    'no_of_dependents': 2,
    'education': 1,  # Assuming 1 for "Graduate", 0 for "Not Graduate"
    'self_employed': 0,  # Assuming 1 for "Yes", 0 for "No"
    'income_annum': 50000,
    'loan_amount': 120000,
    'loan_term': 360,
    'cibil_score': 0.75,
    'customerAssets': 0.85
}

loan_status = predict_loan_approval(knn_classifier, input_data)
print(f"Loan status: {loan_status}")
```
